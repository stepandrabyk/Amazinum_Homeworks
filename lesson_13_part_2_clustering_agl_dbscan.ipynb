```python
from sklearn.cluster import KMeans
```

```python
J=[]
for number_claster in range(3,20):
    clf= KMeans(n_clusters=number_claster,n_init=10).fit(X)
    J_best=clf.inertia_
    J.append([number_claster,J_best])
print(J)
J = np.array(J)
```

```python
plt.figure ()
plt.plot(J[:,0],J[:,1])
plt.xlabel('Iterations')
plt.ylabel('Cost function J')
plt.scatter(J[:,0],J[:,1])
```

## The best number of clusters is 5 by Elbow method

```python
clf= KMeans(n_clusters=5,n_init=10).fit(X)
print ('Cenroids:')
centroids = clf.cluster_centers_  
predicted = clf.predict(X)
plt.figure()
colors = np.array(['green','orange','grey', 'brown','yellow','black'])

# draw samples 
plt.scatter(X[:,0],X[:,1], c= colors[predicted])

# draw centroids 
for i, c in enumerate (clf.cluster_centers_):    
    plt.plot(c[0], c[1], marker = 'x', color= colors[i+1], markersize=14)
```
